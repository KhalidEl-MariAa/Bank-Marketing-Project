# Convert to numerical data

```python
obj = data.select_dtypes(include = "object")
obj.columns
```

```python
Encoder = skp.LabelEncoder()
for col in obj.columns:
    data[col] = Encoder.fit_transform(data[col])

```

# Scaling

```python
Y= data["y"]
X = data.drop("y",axis=1)
print(Y.shape)
print(X.shape)
X.head(5)
```

```python
scaled_columns = []

for col in X.columns:
    if col in list(obj.columns):
        continue
    else:
        scaled_columns.append(col)
print(obj.columns)
scaled_columns
```

```python
Scaler = skp.StandardScaler()


scaled = Scaler.fit_transform(X)

X = pd.DataFrame(scaled,columns=X.columns)

X.head(5)
```

```python
ytest.shape
```

# Split Data

```python
xtrain,xtest,ytrain,ytest = skm.train_test_split(X,Y,test_size=0.2,random_state=20)
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

# Creating Model

```python
model = tfk.models.Sequential()
model.add(tfk.layers.Flatten(input_shape=(20,)))
model.add(layer = tfk.layers.Dense(units = 64,activation= tfk.activations.relu))
model.add(layer = tfk.layers.Dense(units = 32, activation= tfk.activations.relu ))
model.add(layer = tfk.layers.Dense(units = 16, activation= tfk.activations.relu ))
model.add(layer = tfk.layers.Dense(units = 1, activation= tfk.activations.sigmoid,))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
model.fit(xtrain,ytrain,batch_size=10,epochs = 5,)
```

```python
ypredict = model.predict(xtest)
```

```python
ypredict = np.round(ypredict)
ypredict
```

```python
model.evaluate(xtest,ytest)
```

```python
MSE = met.mean_squared_error(y_true=ytest,y_pred=ypredict)
MSE
```

```python
classification_report = met.classification_report(ytest,ypredict,output_dict=True)
```
